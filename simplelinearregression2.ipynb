```python
#x , y دو روش جداکردن
#drop روش اول با
x=df.drop('cylandr',axis=1)
x=x.drop('fuelcomb',axis=1)
x=x.drop('out1',axis=1)
#drop از قلم انداختن  رها کردن


#روش دوم
#y=df.engine


#روش اول داده هارو تو جدول میده بهمون اما روش دوم داده هارو خارج از جدول میده

y=df.out1
```

```python
x_train , x_test , y_train , y_test=train_test_split(x,y,train_size=0.8)
#ترتیب چهارتا 
#x, y  
#مهمه
```
