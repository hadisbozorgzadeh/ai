```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
model=linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
out_robat=model.predict(x_test)
```

```python
out_robat
```

```python
y_test
```

```python
error=y_test-out_robat
#مقدار خطا
```

```python
error
```

```python
true,false=0,0
for i in error:
    if i==0:
        true+=1
    else:
        false+=1
print(f"true={true}")
print(f"false={false}")
```
