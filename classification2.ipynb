```python
#مسِله اصلی از اینجا شروع میشه
#ساخت دیتاست x
x=np.empty((0,4096))
#ساخت ماتریس خالی 0*4096
for i in range(1,21):
    img=Image.open("image3/"+str(i)+".png")
    p2=np.array(img)
    p2=p2[:,:,0]
    p2=p2.flatten()
    #p2=p2/255
    x=np.append(x,[p2],axis=0)
    #به ایکس سطرهای پی 2 را اضافه میکند
    
#4096 یا 4095
#empty خالی
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
out_robat=model.predict(x_test)
```

```python
out_robat
```

```python
error=y_test - out_robat
```
