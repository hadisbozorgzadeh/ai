```python
import numpy as np
import seaborn as sns;sns.set()
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x,y_true=make_blobs(n_samples=400,centers=4,cluster_std=0.60,random_state=0)
#random_state
```

```python
x 
```

```python
x[0]
#منفی کناره مولفه اولش برا اعشار هست 
#2.5  -1    ==   0.25  
```

```python
y_true
#{0,1,2,3}
```

```python
plt.scatter(x[:,1],x[:,0],s=20)
#های نمودار xبرای محور xستون یکم 
#های نمودار yبرای محور xستون صفرم 

#s=size
#set جزو s
#کار با نمودار و حباب set

plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(x)
out_robat=model.predict(x)
```

```python
out_robat
```

```python
plt.scatter(x[:,1],x[:,0],c=out_robat,cmap="summer",s=30)

#c=center
#cmap= ?
#set جزو cmap و c
```

```python
center=model.cluster_centers_
```

```python
center
```

```python
plt.scatter(center[:,1],center[:,0],s=30)
plt.show()
```

```python
#داده های دسته بندی شده با هوش همراه مرکزها
plt.scatter(x[:,1],x[:,0],c=out_robat,cmap="summer",s=20)
center=model.cluster_centers_
plt.scatter(center[:,1],center[:,0],c="red",s=100,alpha=0.9)
plt.show()
#c=center or color 
#alpha
```

```python

#مقایسه کن
#با بالایی
plt.scatter(x[:,1],x[:,0],c=y_true,cmap="summer",s=20)
center=model.cluster_centers_
plt.scatter(center[:,1],center[:,0],c="red",s=100,alpha=0.9)
plt.show()
```

```python
#مقایسه کن 
#با دو تا بالاییی
plt.scatter(x[:,0],x[:,1],s=20,cmap='summer')
center=model.cluster_centers_
plt.scatter(center[:,0],center[:,1],c="red",s=100,alpha=0.9);
plt.show()
```
