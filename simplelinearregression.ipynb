```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
import sklearn.metrics as sm
```

```python
df=pd.read_csv("co2.csv")
#df = data file 
```

```python
df
```

```python
df.describe()

#std=standard deviationانحراف معیار
```

```python
sb.countplot(x='out1',data=df)
#plot  نمودار یا طرح
#countplot نمودار شمارشی
```

```python
plt.subplots(figsize=(9,9))
sb.heatmap(df.corr(),annot=True)
#sub
#heatmap تاثیر ورودی ها بر خروجی
```

```python
x=df.drop("out1",axis=1)
x=x.drop("engine",axis=1)
x=x.drop("cylandr",axis=1)
y=df.out1
#drop حذف کردن ستونی
#x=fuelcomb
#y=out1
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test =train_test_split(x,y,test_size=0.2)

#split جداکننده
#در پرانتز سایز تست یا سایز ترین را مشخص میکنیم سایز اون یکی هم با مشخص کردن یکیش اتومات حساب میشه
#معمولا :
#test_size=0.2  ,  train_size=0.8
#داده های تست و ترین به صورت رندوم انتخاب می شوند
```

```python
x_test 
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

```python
model=linear_model.LinearRegression()
#نوع رگرسیون یا رگرشن مشخص میکنخ واسه مدل
```

```python
model.fit(x_train,y_train)
```

```python
out_robat=model.predict(x_test)
```

```python
out_robat
```

```python
y_test
```

```python
sm.mean_squared_error(y_test,out_robat)
```
